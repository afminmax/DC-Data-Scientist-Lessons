```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

## Stock Visualization

```python
df = pd.read_csv('4stocks.csv')
```

```python
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
y_columns = ['IBM','GOOG','AAPL','AMZN']
```

```python
#df.plot(figsize=(10,5))
df.plot(x='Date', y=y_columns, figsize=(13,5))
#df.plot(figsize=(10,5))
plt.title('Tech Titans in 2019 YTD')
plt.xlabel('Day')
plt.ylabel('Price USD')
plt.show()
```

## Mile Per Gallon Visualization

```python
mpgdf = pd.read_csv('auto-mpg.csv')
```

```python
mpgdf
```

```python
mpgdf.describe()
```

```python
mpgdf.plot(kind='scatter', x='hp', y='mpg')
plt.title('Fuel efficiency vs Horse-power')
plt.xlabel('Horse-Power')
plt.ylabel('Miles Per Gallon')
```

## Box Plots

```python
mpg_columns = ['weight','mpg']
```

```python
mpgdf.plot(subplots=True, figsize=(13,10))
```

```python
mpgdf[mpg_columns].plot(subplots=True, figsize=(13,10))
```

```python
mpgdf[mpg_columns].plot(kind='box', subplots=True, figsize=(13,10))
```
